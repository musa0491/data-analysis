##Making changes to the data

```
df.head(5)
```

```
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

```
df.head()
```

```
df.sort_values('Total')
```

```
df = df.drop(columns = ['Total'])
```

```
df
```

```
df['Total'] = df.iloc[ : ,4 : 10].sum(axis = 1) #1 for horizontal adding aand 0 for vertical
```

```
df
```

```
df[['Total','HP','Defense','Name','Attack']]
```

```
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
```

##Aggregate Statistics (Group by)

```
df = pd.read_csv('modified_pokemon.csv')
```

```
df.groupby(['Type 1']).mean()
```

```
df.groupby(['Type 2']).mean().sort_values('HP')
```

```
#sum mean count 
```

```
df['count'] = 1
df
```

```
df.groupby(['Type 1']).count()['count']
```
