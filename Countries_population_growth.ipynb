```
countries = pd.read_csv('countries.csv')
countries
```

```
zim = countries[countries.country == 'Zimbabwe']
afg = countries[countries.country == 'Afghanistan']
chi = countries[countries.country == 'China']
usa = countries[countries.country == 'United States']
```

```
plt.plot(zim['year'], zim['population'] / 10 ** 6)
plt.plot(afg['year'], afg['population'] / 10 ** 6)
plt.plot(chi['year'], chi['population'] / 10 ** 6)
plt.plot(usa['year'], usa['population'] /10 ** 6)
plt.title('Population Growth')
plt.xlabel('year')
plt.ylabel('population')
plt.legend(['Zimbabwe','Afghanistan', 'China', 'Unites States'])
plt.show()
```
